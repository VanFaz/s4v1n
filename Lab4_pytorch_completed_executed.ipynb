### Задача 1

1). Создайте два вещественных тензора: `a` размером (3, 4) и `b` размером (12,)   
2). Создайте тензор `c`, являющийся тензором `b`, но размера (2, 2, 3)  
3). Выведите первый столбец матрицы `a` с помощью индексации

```python
# 1) Два вещественных тензора: a размера (3, 4), b размера (12,)
a = torch.arange(12, dtype=torch.float32).reshape(3, 4)
b = torch.arange(12, dtype=torch.float32)

# 2) Тензор c — это b, изменённый до размера (2, 2, 3)
c = b.reshape(2, 2, 3)

# 3) Первый столбец матрицы a
first_column = a[:, 0]

print("a =")
print(a)
print("\nb =")
print(b)
print("\nc =")
print(c)
print("\nПервый столбец a:")
print(first_column)
```

### Задача 2

1). Создайте два вещественных тензора: `a` размером (5, 2) и `b` размером (1,10)   
2). Создайте тензор `c`, являющийся тензором `b`, но размера (5, 2)  
3). Произведите все арифметические операции с тензорами `a` и `c`

```python
# 1) Два вещественных тензора: a размера (5, 2), b размера (1, 10)
a = torch.arange(1, 11, dtype=torch.float32).reshape(5, 2) / 10
b = torch.arange(1, 11, dtype=torch.float32).reshape(1, 10)

# 2) Тензор c — это b, изменённый до размера (5, 2)
c = b.reshape(5, 2)

print("a =")
print(a)
print("\nc =")
print(c)

# 3) Арифметические операции с тензорами a и c
print("\na + c =")
print(a + c)

print("\na - c =")
print(a - c)

print("\na * c =")
print(a * c)

print("\na / c =")
print(a / c)

print("\na ** 2 =")
print(a ** 2)

print("\nc ** 2 =")
print(c ** 2)
```

### Задача 3

1). Создайте тензор *целых чисел* `images` размерности (100, 200, 200, 3) (можно интерпретировать это как 100 картинок размера 200х200 с тремя цветовыми каналами, то есть 100 цветных изображений 200х200). Заполните его нулями  

2). Сделайте так, чтобы у $i$-ой по порядку картинки была нарисована белая полоса толщиной два пикселя в строках, которые соответствуют номеру картинки. Например, у 3-ей по порядку картинки белая полоска будет занимать 6 и 7 строки, у 99 -- 198 и 199. Сделать белую строку можно, записав в ячейки тензора число 255 (по всем трём каналам). Выведите пару примеров с помощоью `pylot`  

3). Посчитайте среднее тензора `images` по 1-ой оси (по сути - средняя картинка), умножьте полученный тензор на 70. Нарисуйте с помощью `pyplot`, должна получиться серая картинка (при взятии среднего нужно привести тензор к float с помощью `.float()`, при отрисовке обратно к int с помощью `.int()`)

```python
# 1) Тензор целых чисел images размера (100, 200, 200, 3), заполненный нулями
images = torch.zeros((100, 200, 200, 3), dtype=torch.uint8)

# 2) Для i-й картинки рисуем белую полосу толщиной 2 пикселя
#    в строках 2*i и 2*i + 1.
for i in range(images.shape[0]):
    images[i, 2 * i:2 * i + 2, :, :] = 255

# Выведем несколько примеров
example_ids = [3, 50, 99]

fig, axes = plt.subplots(1, len(example_ids), figsize=(12, 4))
for ax, idx in zip(axes, example_ids):
    ax.imshow(images[idx].numpy())
    ax.set_title(f"Картинка {idx}: строки {2*idx}-{2*idx + 1}")
    ax.axis("off")
plt.show()

# 3) Среднее по первой оси: усредняем все 100 изображений, затем умножаем на 70
mean_image = images.float().mean(dim=0) * 70

plt.figure(figsize=(5, 5))
plt.imshow(mean_image.int().numpy())
plt.title("Средняя картинка * 70")
plt.axis("off")
plt.show()
```

### Задача 4

Перейдите на Google Colaboratory, чтобы работать с GPU. Загрузите туда этот ноутбук (Upload notebook)
В Colab нажмите в меню сверху: Runtime -> Change Runtime Type -> GPU, где изначально стоит None
Объявите тензор a размера (2, 3, 4) и тензор b размера (1, 8, 3) на GPU, иницилизируйте их случайно равномерно (.uniform_())
Затем измените форму тензора b, чтобы она совпадала с формой тензора a, получите тензор c
Переместите c на CPU, переместите a на CPU
Оберните их в Variable()
Объявите тензор L = torch.mean((c - a) ** 2) и посчитайте градиент L по c ( то есть  ∂𝐿∂𝑐)
 
Выведите градиент L по c

```python
# В Colab предварительно нужно включить GPU:
# Runtime -> Change Runtime Type -> GPU.
# Код ниже использует GPU, если он доступен; иначе выполнится на CPU,
# чтобы ячейку можно было проверить и вне Colab.

from torch.autograd import Variable

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("Используем устройство:", device)

# Объявляем тензоры на выбранном устройстве и инициализируем равномерно
a_gpu = torch.empty((2, 3, 4), device=device).uniform_()
b_gpu = torch.empty((1, 8, 3), device=device).uniform_()

# Меняем форму b так, чтобы она совпала с формой a
c_gpu = b_gpu.reshape_as(a_gpu)

# Перемещаем a и c на CPU
a_cpu = a_gpu.cpu()
c_cpu = c_gpu.cpu()

# Оборачиваем в Variable; градиент нужен только по c
a = Variable(a_cpu)
c = Variable(c_cpu, requires_grad=True)

# L = mean((c - a)^2), затем считаем dL/dc
L = torch.mean((c - a) ** 2)
L.backward()

print("L =", L.item())
print("Градиент dL/dc:")
print(c.grad)
```

### Задача 5

Доработайте многослойную нейронную сеть из файла "4_3 mlp pytorch tasks".

```python
# Доработка многослойной сети из файла «4_3 mlp pytorch tasks».
# Используем тот же датасет-«спираль», но берём более глубокую MLP,
# нелинейность Tanh и оптимизатор Adam, чтобы граница классов стала лучше.

import numpy as np
from torch import nn

torch.manual_seed(42)
np.random.seed(42)

# Ограничим число потоков: на маленьком датасете так обучение стабильнее и быстрее.
torch.set_num_threads(1)

# Генерация датасета, как в файле 4_3
N = 100
D = 2
K = 3

X_np = np.zeros((N * K, D), dtype=np.float32)
y_np = np.zeros(N * K, dtype=np.int64)

for j in range(K):
    ix = range(N * j, N * (j + 1))
    r = np.linspace(0.0, 1, N)
    t = np.linspace(j * 4, (j + 1) * 4, N) + np.random.randn(N) * 0.2
    X_np[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
    y_np[ix] = j

plt.figure(figsize=(6, 5))
plt.scatter(X_np[:, 0], X_np[:, 1], c=y_np, s=40, cmap=plt.cm.rainbow)
plt.title("Исходный датасет: спираль")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()

# Оборачиваем данные в torch-тензоры
X = torch.autograd.Variable(torch.FloatTensor(X_np))
y = torch.autograd.Variable(torch.LongTensor(y_np))

# Улучшенная архитектура: несколько скрытых слоёв вместо одного
improved_net = torch.nn.Sequential(
    torch.nn.Linear(2, 64),
    torch.nn.Tanh(),
    torch.nn.Linear(64, 64),
    torch.nn.Tanh(),
    torch.nn.Linear(64, 3),
)

loss_fn = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(improved_net.parameters(), lr=1e-2, weight_decay=1e-4)

loss_history = []
accuracy_history = []

for epoch in range(501):
    # forward
    y_pred = improved_net(X)
    loss = loss_fn(y_pred, y)

    # backward + update
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # контроль качества
    with torch.no_grad():
        preds = torch.argmax(improved_net(X), dim=1)
        accuracy = (preds == y).float().mean().item()

    loss_history.append(loss.item())
    accuracy_history.append(accuracy)

    if epoch % 100 == 0:
        print(f"epoch={epoch:3d} | loss={loss.item():.4f} | accuracy={accuracy:.3f}")

print(f"Финальная accuracy: {accuracy_history[-1]:.3f}")

plt.figure(figsize=(6, 4))
plt.plot(loss_history)
plt.title("Loss во время обучения")
plt.xlabel("epoch")
plt.ylabel("CrossEntropyLoss")
plt.show()

# Отрисовка разделяющей поверхности
h = 0.02
x_min, x_max = X_np[:, 0].min() - 1, X_np[:, 0].max() + 1
y_min, y_max = X_np[:, 1].min() - 1, X_np[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
grid_tensor = torch.FloatTensor(np.c_[xx.ravel(), yy.ravel()])

with torch.no_grad():
    Z = improved_net(grid_tensor)
    Z = torch.argmax(Z, dim=1).numpy()
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, cmap=plt.cm.rainbow, alpha=0.3)
plt.scatter(X_np[:, 0], X_np[:, 1], c=y_np, s=40, cmap=plt.cm.rainbow)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title("Улучшенная MLP для классификации спирали")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
